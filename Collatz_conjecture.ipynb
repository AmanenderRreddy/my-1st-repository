```
#Collatz conjecture
for i in range(1,100000):
  n=i
  print("number is",n)
  while n!=1:
    if(n%2==0):
      print(n,"is a even number so divide it by 2")
      n=n/2
      if(n==1):
        print("n is 1")
        print("Collatz conjecture is success")
        break
    if(n%2!=0):
      print(n,"is a odd number so multiply it by 3 and add 1 ")
      n=(n*3)+1
      if(n==1):
        print("n=1")
        print("Collatz conjecture is success")
        break
  


     
```

```
import io
import pandas as pd
import numpy as np
df2 = pd.read_csv(io.BytesIO(uploaded['dataset_3.csv']))
data = df2.values

# Loading A and B 
A = data[:, 0]
B = data[:, 1]
```

```

def distance(a, b):
    # Magnitude of the difference vector is
    # the distance between two points (a and b)
    return np.linalg.norm(a - b)
```

```


def normalize(x):
    # Shift all the values in x in the range of [0, 1]
    num = x - np.min(x)
    den = np.max(x) - np.min(x)
    return (num / den)
```

```



def closest_mean(x, means):
    # Calculate the closest mean from the point 'x'
    k = len(means)
    d = float('inf')
    idx = -1
    for i in range(k):
        temp = distance(means[i], x)
        if temp < d:
            d = temp
            idx = i
    return idx
```

```
# Plot dataset using matplotlib
import matplotlib.pyplot as plt

plt.plot(A, B, '.')
plt.title('B and A')
plt.xlabel('A')
plt.ylabel('B')
plt.show()
```

```
# Normalize the A and B
A = normalize(A)
B = normalize(B)
plt.plot(A, B, '.')
plt.title('Normalized B and A')
plt.xlabel('A')
plt.ylabel('B')
plt.show()
```

```
# Initialize random means (or centroids)
K = 3
means = []
for i in range(K):
    x = np.random.rand()
    y = np.random.rand()
    mean = [x, y]
    means.append(mean)

means = np.array(means)

plt.plot(A, B, '.')
plt.plot(means[0][0], means[0][1], 'ro')
plt.plot(means[1][0], means[1][1], 'go')
plt.plot(means[2][0], means[2][1], 'yo')
plt.title('Initializing random means')
plt.xlabel('A')
plt.ylabel('B')
plt.show()
```

```
# Assign labels to each of the point in the dataset
l = len(B)
labels = [0] * l
for i in range(l):
    A1 = A[i]
    B1= B[i]
    x = [A1, B1]
    x = np.array(x)
    label = closest_mean(x, means)
    labels[i] = label

labels = np.array(labels)
a, b = A[labels == 0], B[labels == 0]
c, d = A[labels == 1], B[labels == 1]
e, f = A[labels == 2], B[labels == 2]

plt.plot(a, b, 'r.')
plt.plot(c, d, 'g.')
plt.plot(e, f, 'y.')
plt.plot(means[0][0], means[0][1], 'rX', markersize=15)
plt.plot(means[1][0], means[1][1], 'gX', markersize=15)
plt.plot(means[2][0], means[2][1], 'yX', markersize=15)
plt.plot()
```

```

# Update means according to the points assigned in the previous steps

means[0] = [a.mean(), b.mean()]
means[1] = [c.mean(), d.mean()]
means[2] = [e.mean(), f.mean()]
plt.plot(a, b, 'r.')
plt.plot(means[0][0], means[0][1], 'rX', markersize=15)

plt.plot(c, d, 'g.')
plt.plot(means[1][0], means[1][1], 'gX', markersize=15)

plt.plot(e, f, 'y.')
plt.plot(means[2][0], means[2][1], 'yX', markersize=15)

plt.title('K means (reassigning labels)')

plt.show()

# Calculate nearest mean
# Store the label of the nearest mean
# Update the mean
```
